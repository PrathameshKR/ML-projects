Importing Dependenices

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and Data processing

```python
sonar_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Copy of sonar data.csv', header=None)#we dont have named attributes hence we put header as none(it is neccessary to do so when u dont have named atributes)
sonar_data.head()
```

```python
#checking the no. of tuples in dataset
sonar_data.shape
```

```python
#Checking if null values present
sonar_data.isnull().sum()
```

```python
#checking if dataset is balanced
sonar_data[60].value_counts()
```

```python
#Checking if values for rock and mine are actually differentiable
sonar_data.groupby(60).mean()
```

```python
#separating labels from dataset
X = sonar_data.drop(columns=60,axis=1)
Y = sonar_data[60]
```

```python
#Creating training and testing datasets
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training --> Logistic Regression

```python
model = LogisticRegression()
#training the logsitic regression model with training data
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#checking accuracy on training data
X_train_prediciton = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediciton,Y_train)
print('Accuracy on training data : ' ,training_accuracy)
```

```python
#Checking accuracy on testing data
X_test_prediciton = model.predict(X_test)
training_accuracy = accuracy_score(X_test_prediciton,Y_test)
print('Accuracy on tesr data : ' ,training_accuracy)
```
