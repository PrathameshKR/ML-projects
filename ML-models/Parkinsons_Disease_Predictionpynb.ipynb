```python
#Importing Dependencies
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import warnings

warnings.filterwarnings("ignore")
```

```python
#Data Collection and Preprocessing
parkinsons_data = pd.read_csv('/content/parkinsons.csv')
```

```python
parkinsons_data.head()
```

```python
#checking no. of columns and rows
parkinsons_data.shape
```

```python
#checking for missing values
parkinsons_data.info()
```

```python
#checking statistical measures of dataset
parkinsons_data.describe()
```

```python
#Checking distribution of target variable
parkinsons_data['status'].value_counts()
```

1.  0 ---> parkinsons positive
2.  1 ---> parkinsons neagative




```python
#Separating features and target
X = parkinsons_data.drop(['name','status'],axis=1)
Y = parkinsons_data['status']
```

```python
print(X,Y)
```

```python
#Splitting data into train and test set
#we can standardize dataset either before splitting or after splitting, each has its pros and cons
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Data standardization
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaler.transform(X_train)
scaler.transform(X_test)
```

```python
#Loading Model
model = svm.SVC(kernel='linear')
```

```python
#Training model
model.fit(X_train,Y_train)
```

```python
#Model Evaluation
train_set_prediction = model.predict(X_train)
training_accuracy = accuracy_score(train_set_prediction,Y_train)
print("Training Accuracy: ",training_accuracy)
```

```python
test_set_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(test_set_prediction,Y_test)
print("Testing Accuracy: ",testing_accuracy)
```

```python
#Building a predictive system
input = (116.68200,131.11100,111.55500,0.01050,0.00009,0.00544,0.00781,0.01633,0.05233,0.48200,0.02757,0.03858,0.03590,0.08270,0.01309,20.65100,0.429895,0.825288,-4.443179,0.311173,2.342259,0.332634)

#converting input to numpyn array
input_as_numpy_array = np.asarray(input)

#Reshaping input as we are predicting for only one instance
input_reshaped = input_as_numpy_array.reshape(1,-1)

#standardize data
input_std = scaler.transform(input_reshaped)

prediction = model.predict(input_std)

if (prediction[0] == 0):
  print("Person doesn't have Parkinsons Disease")
else:
  print("Person has Parkinsons Disease")
```
